# The Battle of the Neighbourhoods | Capstone Project Introduction

### Introduction:

This project aims to compare the neighbourhoods accross Scarborough, Toronto in order to objectively select the best neighbourhood for the client's purposes.

A large number of individuals relocate every yeaar and this project will assist them in choosing the best neighbourhood in Toronto, Canada for their needs, whether it be the areas': location; access to a hospital; school; cafe; restaurant; shopping mall and so on.

This project will analyse and compare a number of features for those moving to Scarborough, Toronto. These features include crime rates, emergency management, median housing price, school ratings, recreational facilities,  road connectivity, weather conditions, water resources as well as wastewater conveyed.

Therefore; people's awareness of certain neightbourhoods will rise alllowng them to make better decisions before deciding on which areas to move for housing, work, schooling and more.

### Problems being targeted:

Since this project tries to help individuals moving to a new neighbourhood by objectively comparing areas in a city and deeming the best one, certain data commonly analysed by people migrating will be targeted such as:

- A list of available homes and their houseing prices.
- A list of newaby schools; including their ratings, location and etc.

### The Location:

The location picked  for this will be Scarborough due to its strong multicultural presence and also being a popular area for immigration across Canada.

### Foursquare:
